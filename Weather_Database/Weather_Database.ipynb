```python
# Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import requests
import sys
from citipy import citipy
from datetime import datetime

# Import the API Key
sys.path.insert(0, '..')
from config import weather_api_key
```

```python
# Create a set of randon latitude and longitude combinations.
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lngs = zip(lats, lngs)
```

```python
# Add the latitude and longitude to a list.
coordinates = list(lat_lngs)
```

```python
# Create a list for holding the cities using citipy for each lat and 
# lng. Append unique city name to cities list.
cities = []

for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    if city not in cities:
        cities.append(city)

# Print the count to confirm count.
len(cities)
```

```python
# Starting URL for weather map API call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
```

```python
# Iterating thru cities list 
city_data = []

# Printing the begginning of the logging.
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Counters
record_count = 1
set_count = 1

for i, city in enumerate(cities):
    
    # Group cities in sets of 50
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
    
    # Create endpoint URL w each city.
    city_url = url + "&q=" + city.replace(" ","+")
    
    # Log the URL, record, and set numbers and city.
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    
    # Add to the record count.
    record_count += 1
    
    # API request for each of the cities.
    try:
        # Parse and retrieve JSON data.
        city_weather = requests.get(city_url).json()
        city_lat = city_weather["cood"]["lat"]
        city_lng = city_weather["cood"]["lon"]
        city_desc = city_weather["weather"][0]["description"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        
        # Append city info into city_data list.
        city_data.append({"City": city.title(),
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Current Description": city_desc,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Country": city_country})
        
# If error, skip city.
    except: 
        print("City not found. Skipping...")
        pass

# Printing the ending of the logging.
print("-----------------------------")
print("Data Retrieval Complete.     ")
print("-----------------------------")
```

```python
# Convert array to DataFrame
city_data_df = pd.DataFrame(city_data)
city_data_df.head()
```
